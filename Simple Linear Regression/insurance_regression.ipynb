```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Import 'insurance.csv' file
df = pd.read_csv('insurance.csv')

# Display the first 5 rows of the DataFrame
print(df.head())
```

* This is for my own practice * 
Scatter graph showing age vs. charges using Matplot:

```python
# Use the data in the relevant columns to determine how age affects the insurance cost
# Plot a scatter plot of age vs. insurance cost
x = df.iloc[:,:1].values # Select all rows and the first column (age)
y = df.iloc[:,-1].values # Select all rows and the last column (charges)

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

Scatterplot of age vs. charges using Seaborn:

```python
# Plot a scatter plot of age vs. insurance cost
# Exactly the same as above but using seaborn
sns.scatterplot(data=df, x='age', y='charges', color='red', hue='smoker')
plt.show()
```

```python
# Fit the model 
linear_model = LinearRegression()
linear_model.fit(x,y)
```

```python
# Plot the data and model
y_pred = linear_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

Again, practicing same concept with Seaborn.

```python
sns.regplot(data=df, x='age', y='charges', scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

Predicting the insurance charge for someone who is 40 years old.


```python
unk_x = [[40]]  # New age

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = linear_model.predict(x_pred)

plt.scatter(x, y, color='blue')
plt.plot(x_pred, y_pred, color='red')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs Insurance Charges with Prediction')
plt.show()

prediction = linear_model.predict(unk_x)
print("Predicted insurance charge for age 40: ", round(prediction[0], 2))

```
